```python
outliers=[]
def detect_outliers(data):
    
    threshold=3
    mean = np.mean(data)
    std =np.std(data)
    
    
    for i in data:
        z_score= (i - mean)/std 
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```

```python
def a_range(start, stop, increment):
    while start < stop: # and not math.isclose(start, stop): Py>3.5
        yield start
        start += increment



import xgboost
from sklearn.model_selection import RandomizedSearchCV
regressor=xgboost.XGBRegressor()
booster=['gbtree','gblinear','dart']

base_score=[ele for ele in a_range(0,1.01,0.05)]
## Hyper Parameter Optimization


n_estimators = [ele for ele in range(5,35)]
max_depth = [ele for ele in range(1,26)]
# booster=['gbtree','gblinear']
learning_rate=[ele for ele in a_range(0.18,0.3,0.02)]
min_child_weight=[ele for ele in range(1,40)]
colsample_bytree = [ele for ele in a_range(0.3,0.7,0.1)]
alpha = [ele for ele in range(10,30)]

# Define the grid of hyperparameters to search
hyperparameter_grid = {
    'n_estimators': n_estimators,
    'max_depth':max_depth,
    'learning_rate':learning_rate,
    'min_child_weight':min_child_weight,
    'booster':booster,
    'base_score':base_score,
    'colsample_bytree' : colsample_bytree ,
    'alpha':alpha
    }
# Set up the random search with 4-fold cross validation
random_cv = RandomizedSearchCV(estimator=regressor,
            param_distributions=hyperparameter_grid,
            cv=5, n_iter=50,
            scoring = 'neg_mean_absolute_error',n_jobs = 4,
            verbose = 5, 
            return_train_score = True,
            random_state=42)
random_cv.fit(X_train,y_train)
```
